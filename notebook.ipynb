```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Checking that the data is imported successfully
netflix_df.head()
```

```python
# Filtering Netflix dataframe to get the movies only.
netflix_movies_df = netflix_df[netflix_df['type'] == 'Movie']
netflix_movies_df.head()
```

```python
# Filtering Netflix dataframe to only have the 90s movies.
netflix_movies_90s_df = netflix_movies_df[netflix_movies_df['release_year'] >= 1990]
netflix_movies_90s_df = netflix_movies_90s_df[netflix_movies_df['release_year'] < 2000]
netflix_movies_90s_df.head()
```

```python
# Checking movies duration distribution
plt.hist(netflix_movies_90s_df.loc[:,'duration'])
plt.show()
```
